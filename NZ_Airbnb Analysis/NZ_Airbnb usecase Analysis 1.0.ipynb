```python
TRUE_DATE_COLS = [
    'last_scraped',
    'calendar_last_scraped',
    'last_searched', 
    'host_since',
    'first_review',
    'last_review',
]

date_cols_true = [c for c in TRUE_DATE_COLS if c in df.columns]

for c in date_cols_true:
    s = pd.to_datetime(
        df[c].astype(str).str.strip(),
        errors='coerce',
        dayfirst=True,
        utc=False
    )
    try:
        s = s.dt.tz_convert(None)
    except Exception:
        pass
    df[c] = s

print("Parsed date columns & ranges:")
for c in date_cols_true:
    s = df[c]
    vmin, vmax = s.min(), s.max()
    print(f"- {c:<24} -> "
          f"{vmin.date() if pd.notna(vmin) else None}  to  "
          f"{vmax.date() if pd.notna(vmax) else None}  "
          f"(non-null={s.notna().sum()})")
```

```python
df = df.copy()

total_hosts = len(df)
print(f"Total hosts: {total_hosts}")

super_mask = (
    df['host_is_superhost']
      .astype(str)
      .str.strip()
      .str.lower()
      .isin(['t', 'true', '1'])
)
super_df = df.loc[super_mask].copy()

super_hosts = len(super_df)
print(f"Superhosts: {super_hosts}")

super_df['response_rate_raw'] = pd.to_numeric(
    super_df['host_response_rate'],
    errors='coerce'
)

max_val = super_df['response_rate_raw'].max()

if max_val < 1e-3:
    scale = 1e8           
elif max_val <= 1:
    scale = 100          
else:
    scale = 1             


super_df['response_rate_pct'] = super_df['response_rate_raw'] * scale


rates = super_df['response_rate_pct'].dropna()

rate_counts = (
    rates.round(0)                     
         .value_counts()
         .sort_index()
         .rename_axis('response_rate_pct')
         .reset_index(name='count')
)

display(
    rate_counts.style.format({
        'response_rate_pct': '{:.0f}%',
        'count': '{:,.0f}'
    })
)

rates_no_100 = rate_counts[rate_counts['response_rate_pct'] < 100]['response_rate_pct']
rates_expanded = np.repeat(
    rate_counts[rate_counts['response_rate_pct'] < 100]['response_rate_pct'].values,
    rate_counts[rate_counts['response_rate_pct'] < 100]['count'].values
)

rates_series = pd.Series(rates_expanded)

bins = [0, 80, 90, 95, 99]
labels = ['0–80%', '80–90%', '90–95%', '95–99%']

bucket = pd.cut(
    rates_series,
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=True
)

bucket_counts = (
    bucket.value_counts()
          .reindex(labels, fill_value=0)
          .rename_axis('bucket')
          .reset_index(name='count')
)

total = bucket_counts['count'].sum()
bucket_counts['share'] = bucket_counts['count'] / total

print("Response rate buckets among Superhosts (excluding 100%):")
display(
    bucket_counts.style.format({'count': '{:,.0f}', 'share': '{:.2%}'})
)

plt.figure(figsize=(6, 4))

bars = plt.barh(bucket_counts['bucket'], bucket_counts['count'])

plt.xlabel('Number of Superhosts')
plt.ylabel('Response rate bucket')
plt.title('Response rate buckets (excluding 100%)')

for bar in bars:
    width = bar.get_width()
    y_pos = bar.get_y() + bar.get_height() / 2
    plt.text(
        width,
        y_pos,
        f'{int(width)}',
        va='center',
        ha='left',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
df = df.copy()

super_mask = (
    df['host_is_superhost']
      .astype(str)
      .str.strip()
      .str.lower()
      .isin(['t', 'true', '1'])
)
super_df = df.loc[super_mask].copy()

print(f"Total hosts: {len(df)}")
print(f"Superhosts: {len(super_df)}")

resp_time_str = (
    super_df['host_response_time']
      .astype('string')      # pandas string dtype
      .str.strip()
      .str.lower()
)

label_map = {
    'within an hour': 'Within an hour',
    'within a few hours': 'Within a few hours',
    'within a day': 'Within a day',
    'a few days or more': 'A few days or more'
}

super_df['host_response_time_label'] = resp_time_str.map(label_map)

super_df['host_response_time_label'] = super_df['host_response_time_label'].fillna('Other / Unknown')


time_counts = (
    super_df['host_response_time_label']
      .value_counts(dropna=False)
      .rename_axis('host_response_time')
      .reset_index(name='count')
)

total_super = len(super_df)
time_counts['share'] = time_counts['count'] / total_super

print("\nHost response time distribution among Superhosts (table):")
display(
    time_counts.style.format({
        'count': '{:,.0f}',
        'share': '{:.2%}'
    })
)

plot_data = time_counts[time_counts['host_response_time'] != 'Within an hour'].copy()

plt.figure(figsize=(6, 4))

x = plot_data['host_response_time']
y = plot_data['count']


bars = plt.bar(x, y)

plt.xlabel('Host response time')
plt.ylabel('Number of Superhosts')
plt.title('Host response time among Superhosts (excluding "Within an hour")')

plt.xticks(rotation=20)


for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  
        height,                             
        f'{int(height)}',                   
        ha='center', va='bottom',          
        fontsize=7
    )

plt.tight_layout()
plt.show()
```

```python
df = df.copy()
df_acc = pd.read_csv("listings.csv", encoding="cp1252")

total_hosts = len(df_acc)
print(f"Total hosts: {total_hosts}")

super_mask = (
    df_acc['host_is_superhost']
        .astype(str)
        .str.strip()
        .str.lower()
        .isin(['t', 'true', '1'])
)
super_df = df_acc.loc[super_mask].copy()

super_hosts = len(super_df)
print(f"Superhosts: {super_hosts}")

acc_str = (
    super_df['host_acceptance_rate']
        .astype(str)
        .str.strip()
)

acc_str = (
    acc_str
        .str.replace('%', '', regex=False)
        .replace({'nan': np.nan, '': np.nan, 'NA': np.nan, 'N/A': np.nan})
)


acc_num = pd.to_numeric(acc_str, errors='coerce')


max_val = acc_num.max()
if pd.notna(max_val) and max_val <= 1.1:
    acc_num = acc_num * 100

super_df['accept_rate_pct'] = acc_num

accept_rates = super_df['accept_rate_pct'].dropna()
print(f"Superhosts with valid acceptance rate: {len(accept_rates)}")
```

```python
bins = [0, 80, 90, 95, 99, 100]
labels = ['0–80%', '80–90%', '90–95%', '95–99%', '100%']

accept_bucket = pd.cut(
    accept_rates,
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=True
)

accept_bucket_counts = (
    accept_bucket.value_counts()
                 .reindex(labels, fill_value=0)
                 .rename_axis('bucket')
                 .reset_index(name='count')
)

total_valid = len(accept_rates)
accept_bucket_counts['share'] = accept_bucket_counts['count'] / total_valid

print("\nAcceptance rate buckets among Superhosts:")
display(
    accept_bucket_counts.style.format({
        'count': '{:,.0f}',
        'share': '{:.2%}'
    })
)

plt.figure(figsize=(6, 4))

x = accept_bucket_counts['bucket']
y = accept_bucket_counts['count']


bars = plt.bar(x, y)

plt.xlabel('Acceptance rate bucket')
plt.ylabel('Number of Superhosts')
plt.title('Acceptance rate buckets among Superhosts')


for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  
        height,                             
        f'{int(height)}',                   
        ha='center', va='bottom',           
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
df = df.copy()

df['host_is_superhost_bin'] = (
    df['host_is_superhost']
      .astype(str)
      .str.strip()
      .str.lower()
      .map({'t': 1, 'true': 1, '1': 1,
            'f': 0, 'false': 0, '0': 0})
      .astype('Int8')
)

df['instant_bookable_bin'] = (
    df['instant_bookable']
      .astype(str)
      .str.strip()
      .str.lower()
      .map({'t': 1, 'true': 1, '1': 1,
            'f': 0, 'false': 0, '0': 0})
      .astype('Int8')
)


label_map = {0: 'Non-Superhost', 1: 'Superhost'}

tmp = df[
    df['host_is_superhost_bin'].notna() &
    df['instant_bookable_bin'].notna()
].copy()

summary_inst = (
    tmp.assign(group=lambda x: x['host_is_superhost_bin'].map(label_map))
       .groupby('group')['instant_bookable_bin']
       .agg(
           n_hosts='count',          
           n_instant='sum',            
           share_instant='mean'        
       )
       .reset_index()
)

print("Instant bookable by Superhost status:")
display(
    summary_inst.style.format({
        'share_instant': '{:.2%}',
        'n_hosts': '{:,.0f}',
        'n_instant': '{:,.0f}'
    })
)


plt.figure(figsize=(5, 4))

x = summary_inst['group']
y = summary_inst['share_instant']

bars = plt.bar(x, y)

plt.ylabel('Share instant-bookable')
plt.title('Instant bookable share by Superhost status')


for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.1%}',
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.ylim(0, 1)
plt.tight_layout()
plt.show()

sup = df[
    (df['host_is_superhost_bin'] == 1) &
    df['instant_bookable_bin'].notna()
].copy()

inst_counts = (
    sup['instant_bookable_bin']
       .value_counts()
       .rename(index={0: 'Not instant bookable', 1: 'Instant bookable'})
       .rename_axis('instant_bookable')
       .reset_index(name='count')
)

inst_counts['share'] = inst_counts['count'] / inst_counts['count'].sum()

print("\nInstant bookable distribution within Superhosts:")
display(
    inst_counts.style.format({
        'count': '{:,.0f}',
        'share': '{:.2%}'
    })
)

plt.figure(figsize=(5, 3))

bars2 = plt.barh(inst_counts['instant_bookable'], inst_counts['count'])

plt.xlabel('Number of Superhost listings')
plt.title('Instant bookable within Superhosts')

for bar in bars2:
    width = bar.get_width()
    y_pos = bar.get_y() + bar.get_height() / 2
    plt.text(
        width,
        y_pos,
        f'{int(width)}',
        va='center',
        ha='left',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```
