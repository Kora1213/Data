```python
def deduplication_better(x):
    trim = x.strip()
    index = (trim+trim).find(trim,1)
    if index != -1:
        return trim[:index]
    else:
        return 'unknown'
```

```python
df['location'] = df['location'].apply(deduplication_better)
```

```python
df['Posted time']=df['Posted time'].astype('string')
```

```python
def apply_posted_time(x):
    if 'd' in x:
        return -int(x[:-1])
    if 'm' in x:
        return -(int(x[:-1])*30)
    else:
        return 0
```

```python
df['Posted time']=df['Posted time'].apply(apply_posted_time)
```
